```python
import pandas as pd
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
labels = cancer['diagnosis']
```

```python
labels.head()
```

```python
features = cancer.drop(['id','diagnosis','Unnamed: 32'], axis=1)
```

```python
features.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features,labels, train_size=0.7, random_state=2529)
```

```python
features_train.shape, features_test.shape, labels_train.shape, labels_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(features_train, labels_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
features_test.head()
```

```python
labels_pred = model.predict(features_test)
```

```python
labels_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(labels_test, labels_pred)
```
